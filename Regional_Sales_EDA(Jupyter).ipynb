```python
import pandas as pd
import plotly.express as px

# Create Year-Month for grouping
df['year_month'] = df['order_date'].dt.to_period('M').astype(str)

# Group by month and sum revenue
monthly_sales = df.groupby('year_month', as_index=False)['revenue'].sum()
monthly_sales = monthly_sales.sort_values('year_month')

# Plot the chart
fig = px.line(
    monthly_sales,
    x='year_month',
    y='revenue',
    markers=True,
    title='Monthly Sales Trend (Revenue)',
    labels={'year_month': 'Month-Year', 'revenue': 'Total Revenue ($)'},
    template='plotly_white',
    height=650
)

fig.update_layout(
    xaxis_title='Month',
    yaxis_title='Revenue (USD)',
    hovermode='x unified',
    title_x=0.5,
    xaxis_tickangle=-45
)

fig.update_yaxes(tickprefix='$', tickformat=',')

fig.show()
```

```python
# 1. Ensure order_date is datetime
df['order_date'] = pd.to_datetime(df['order_date'], dayfirst=True, errors='coerce')

# 2. Remove Jan & Feb 2018
df = df[~((df['order_date'].dt.year == 2018) & 
          (df['order_date'].dt.month.isin([1, 2])))]

# 3. Drop invalid dates (recommended)
df = df.dropna(subset=['order_date'])

# 4. Create month columns (THIS WAS MISSING)
df['month_num'] = df['order_date'].dt.month
df['month_name'] = df['order_date'].dt.month_name()

# 5. Average revenue per calendar month
monthly_avg = df.groupby(['month_num', 'month_name'], as_index=False)['revenue'].mean()
monthly_avg = monthly_avg.sort_values('month_num')

# 6. Plot
import plotly.express as px

fig2 = px.line(
    monthly_avg,
    x='month_name',
    y='revenue',
    markers=True,
    title='Average Monthly Sales Seasonality (All Years Combined)',
    labels={'month_name': 'Month', 'revenue': 'Average Revenue ($)'},
    template='plotly_white',
    height=650
)

fig2.update_layout(
    xaxis_title='Calendar Month',
    yaxis_title='Average Revenue (USD)',
    title_x=0.5,
    yaxis=dict(tickprefix='$', tickformat=',')
)

fig2.update_traces(line=dict(width=3))
fig2.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Data
chan_sales = df.groupby('channel')['revenue'].sum().sort_values(ascending=False)

# Colors
colors = sns.color_palette('coolwarm', len(chan_sales))

# Plot
plt.figure(figsize=(7,7))

wedges, texts, autotexts = plt.pie(
    chan_sales.values,
    labels=chan_sales.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=colors,
    wedgeprops={'edgecolor': 'black', 'linewidth': 1},
    textprops={'fontsize': 11}
)

# Make percentage text bold
for autotext in autotexts:
    autotext.set_color('black')
    autotext.set_fontsize(11)
    autotext.set_weight('bold')

# Add center circle (donut style 🔥)
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
plt.gca().add_artist(centre_circle)

# Title
plt.title('Total Sales by Channel', fontsize=14, weight='bold')

plt.tight_layout()
plt.show()
```

***Top 10 States by Order Count***

```python
# Aggregate order count
state_orders = (
    df.groupby('state')['order_number']
      .nunique()
      .sort_values(ascending=False)
      .head(10)
      .reset_index()
)

# Plot
plt.figure(figsize=(12,6))

sns.barplot(
    data=state_orders,
    x='state',
    y='order_number',
    palette='viridis'
)

plt.title('Top 10 States by Order Count', fontsize=14)
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Calculate average profit margin
channel_profit = (
    df.groupby('channel')['profit_margin_pct']
      .mean()
      .sort_values(ascending=False)
      .reset_index()
)

# Plot
plt.figure(figsize=(8,5))

ax = sns.barplot(
    data=channel_profit,
    x='channel',
    y='profit_margin_pct',
    palette='coolwarm'
)

# Add values on top of bars
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.1f}%",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center',
        va='bottom',
        fontsize=10
    )

# Styling
plt.title('Average Profit Margin by Channel', fontsize=14)
plt.xlabel('Channel')
plt.ylabel('Average Profit Margin (%)')
plt.xticks(rotation=30)

plt.tight_layout()
plt.show()
```

```python
# Aggregate revenue by customer
customer_rev = (
    df.groupby('customer_name')['revenue']
      .sum()
      .sort_values(ascending=False)
      .reset_index()
)

# Top 10
top_10 = customer_rev.head(10)

# Bottom 10
bottom_10 = customer_rev.tail(10).sort_values('revenue', ascending=True)

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(18,6))

# Top 10 plot
sns.barplot(
    data=top_10,
    x='revenue',
    y='customer_name',
    palette='Greens_r',
    ax=axes[0]
)

axes[0].set_title('Top 10 Customers by Revenue')
axes[0].set_xlabel('Revenue')
axes[0].set_ylabel('Customer_name')

# Add labels
for p in axes[0].patches:
    axes[0].annotate(
        f"{p.get_width():,.0f}",
        (p.get_width(), p.get_y() + p.get_height()/2),
        ha='left',
        va='center'
    )

# Bottom 10 plot
sns.barplot(
    data=bottom_10,
    x='revenue',
    y='customer_name',
    palette='Reds',
    ax=axes[1]
)

axes[1].set_title('Bottom 10 Customers by Revenue')
axes[1].set_xlabel('Revenue')
axes[1].set_ylabel('Customer')

# Add labels
for p in axes[1].patches:
    axes[1].annotate(
        f"{p.get_width():,.0f}",
        (p.get_width(), p.get_y() + p.get_height()/2),
        ha='left',
        va='center'
    )

plt.tight_layout()
plt.show()
```

```python

# Aggregate metrics per customer
cust_summary = df.groupby('customer_name').agg(
    total_revenue=('revenue', 'sum'),
    total_profit=('profit', 'sum'),
    avg_margin=('profit_margin_pct', 'mean'),
    orders=('order_number', 'nunique')
)

# Convert revenue to millions
cust_summary['total_revenue_m'] = cust_summary['total_revenue'] / 1e6

plt.figure(figsize=(7, 5))

# Bubble chart with revenue in millions
sns.scatterplot(
    data=cust_summary,
    x='total_revenue_m',        # <-- use revenue in millions
    y='avg_margin',
    size='orders',
    sizes=(20, 200),
    alpha=0.7
)

plt.title('Customer Segmentation: Revenue vs. Profit Margin')
plt.xlabel('Total Revenue (Million USD)')  # <-- updated label
plt.ylabel('Avg Profit Margin (%)')

plt.tight_layout()
plt.show()
```
